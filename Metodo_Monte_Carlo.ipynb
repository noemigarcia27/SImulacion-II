# MÃ©todo de Monte Carlo
**a) Acierto y error **

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd

def g(x):
  return np.sqrt(np.arctan(x))

print(g(1))
aciertos=0
N=1000

X=np.linspace(0,1,100)
Y=g(X)
plt.plot(X,Y)
for i in range(N):  #generamos los pares aleatorios
  x=rd.random()
  y=rd.random()
  if y<=g(x): #cuantos caen debajo de la curva
    plt.scatter(x,y,color="blue")
    aciertos+=1 #vamos contando los puntos que caen debajo de la curva
  else:
    plt.scatter(x,y,color="red")

aciertos

I=aciertos/N #el numero de aciertos entre el total de los puntos
```

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
```

```python
def g(x):
  return np.sqrt(np.arctan(x))
```

```python
def estimacion(N):
  lista=[]
  aciertos=0
  X=np.linspace(0,1,100)
  Y=g(X)
  plt.plot(X,Y)
  for i in range(N):
    x = rd.random()
    y = rd.random()
    if y <= g(x):
      plt.scatter(x,y,color="blue")
      aciertos += 1
    else:
      plt.scatter(x,y,color="red")
  I=aciertos/i+1
  lista.append(I)
  return lista
```

```python
L=estimacion(1000)
```

```python
plt.plot(L)
```

# MÃ©todo Monte Carlo Crudo

```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd
```

La funcion que vamos a utilizar es
g(x) = âarctan(x) dx



```python
def g(x):
  return np.sqrt(np.arctan(x))
```

Paso 1 y Paso 3. (el paso 2 no es necesario porque la integral va de 0 a 1 )

```python
def mc_crudo(N):
  lista=[]
  for i in range(N):
    u=rd.random()
    lista.append(g(u))
  return np.mean(lista)
print(mc_crudo(10))
```

```python
L=[]
for i in range(1000):
  x=mc_crudo(i)
  L.append(x)
```

```python
plt.plot(L)
```
