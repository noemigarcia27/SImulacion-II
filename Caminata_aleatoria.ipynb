```python
import random as rd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
N=10000

y=[0]
x=[0]
z=[0]
```

1d

```python
posiciones=[0]

for i in range (1, N+1):
  paso = rd.choice([-1,1])
  nueva_posicion = posiciones[-1] + paso
  posiciones.append(nueva_posicion)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(posiciones)
```

2d

```python
pasos_x = [rd.choice([-1, 0, 1]) for _ in range(N)]
pasos_y = [rd.choice([-1, 0, 1]) for _ in range(N)]

x = np.cumsum(pasos_x)
y = np.cumsum(pasos_y)

plt.figure(figsize=(10, 6))
plt.plot(x, y)
```

3d

```python
for i in range(N):
  dx = rd.choice([-1, 0, 1])
  dy = rd.choice([-1, 0, 1])
  dz = rd.choice([-1, 0, 1])

  x.append(x[i] + dx)
  y.append(y[i] + dy)
  z.append(z[i] + dz)
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z)
```
