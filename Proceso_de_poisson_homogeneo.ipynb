```python
from scipy.stats import poisson
import matplotlib.pyplot as plt
import numpy as np
import random as rd
```

```python
mu = 4
k_= np.arange(0,12)
probabilidad = []
```

```python
for k in k_:
  probabilidad.append(poisson.pmf(k, mu))
```

```python
plt.stem(k_, probabilidad, basefmt=" ")
plt.xlabel('k (Número de ocurrencias)')
plt.ylabel('Probabilidad P(k)')
plt.grid(True)
plt.show()
```

```python
lambda_ = 4
T=10
t=0
N=100
eventos = [0]
tiempos = [0]
```

```python
for i in range (T):
  U = rd.random()
  t = t - np.log(U) / lambda_

  if t > T:
    break
  else:
    N = N+1
    tiempos.append(t)
    eventos.append(i+1)

tiempos = np.array(tiempos)
eventos = np.array(eventos)
```

```python
plt.step(tiempos, eventos, where="post", color="b", label="Tiempos de eventos")
plt.xlabel('Número de evento (N)')
plt.ylabel('Tiempo t')
plt.title(f"Simulación de Proceso Poisson con λ={lambda_} y T={T}")
plt.grid(True)
plt.show()
```
